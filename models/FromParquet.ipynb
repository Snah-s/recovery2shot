# De parquet a csv
Aqui vamos a leer el archivo .parquet que contiene la data de los eventos y lo convertiremos a un csv


## Imports y configuraci√≥n de los paths

```python
import os
from pathlib import Path
import pandas as pd
BASE_DIR = Path().resolve()
# Path al input Parquet file
PARQUET_PATH = BASE_DIR / ".." / "data" / "source" / "events_all_37_90.parquet"
PARQUET_PATH = PARQUET_PATH.resolve()
OUTPUT_DIR = BASE_DIR / "source"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
# Path  output del CSV 
OUTPUT_CSV = OUTPUT_DIR / "events_all_37_90.csv"
print("Directorio Base", BASE_DIR)
print("RLeyendo desde el Parquet path:", PARQUET_PATH)
print("El csv estara guardado en :", OUTPUT_CSV)
```

## Leyendo el .parquet

```python
df = pd.read_parquet(PARQUET_PATH)
print("DataFrame shape:", df.shape)
df.head()
```

## Guardando el csv

```python

print("Guardando el DataFrame como csv...")
df.to_csv(OUTPUT_CSV, index=False)
print("Archivo convertido y guardado")
```
